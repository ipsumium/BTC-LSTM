```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def periodica_sin(c,t):
    return (150*c[0]*(20*np.sin(c[1]*t+20*c[2])))
def periodica_cos(c,t):
    return (10*c[0]*(40*np.cos(c[1]*t+20*c[2])))
param=[np.random.random(3) for x in range(6)]
t = np.arange(0.0, 300.0, 0.05)
fp1=[periodica_sin(param[0],x) for x in t]
fp2=[periodica_sin(param[1],x) for x in t]
fp3=[periodica_cos(param[2],x) for x in t]
fp4=[periodica_cos(param[3],x) for x in t]
fp5=[periodica_cos(param[4],x) for x in t]
fp6=[periodica_sin(param[5],x) for x in t]
fpt=[sum(x) for x in zip(fp1, fp2,fp3,fp4,fp5,fp6)]
```

```python
fig = plt.figure(figsize=(15, 7))
plt.subplot(2, 3, 1)
plt.plot(t,fp1, label='fp1')
plt.title('fp1')
plt.subplot(2, 3, 2)
plt.plot(t,fp2, label='fp1')
plt.title('fp2')
plt.subplot(2, 3, 3)
plt.plot(t,fp3, label='fp1')
plt.title('fp3')
plt.subplot(2, 3, 4)
plt.plot(t,fp4, label='fp1')
plt.title('fp4')
plt.subplot(2, 3, 5)
plt.plot(t,fp4, label='fp5')
plt.title('fp5')
plt.subplot(2, 3, 6)
plt.plot(t,fp4, label='fp6')
plt.title('fp6')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(15, 7))
plt.plot(t,fpt, label='fpt')
plt.title('fp1+fp2+fp3+fp4+fp5+fp6')
```

```python
time_step = 1 / 30
ps = np.abs(np.fft.fft(fpt))**2
freqs = np.fft.fftfreq(len(fpt), time_step)
idx = np.argsort(freqs)
fig = plt.figure(figsize=(8, 4))
plt.plot(freqs[idx], ps[idx])
plt.tight_layout()
plt.xlim(0,1)
plt.title('Power spectrum')
```
